# Variation 1 — Longest Substring Without Repeating
Given a string s, find the length of the longest substring without repeating characters.

Input s = "abcabcbb" 

Output 3 because abc

Input s = "bbbbb"

Output 1


```python
def length_longest_substring(str):
    seen = set()
    left = 0
    max_len = 0
    for right in range(len(str)):
        while str[right] in seen:
            seen.remove(str[left])
            left += 1
        seen.add(str[right])
        max_len = max(max_len, right-left+1)
    return max_len
```

```python
s = "abcdac"
length_longest_substring(s)
```

Assume n = length of string
m = number of unique characters
- Time complexity: O(n) because each character enters and leaves the window at most once 
- Space complexity: O(m) 

```python
def print_longest_string(s):
    seen = set()
    left = 0
    max_len = 0
    max_str = ""

    for right in range(len(s)):
        while s[right] in seen:
            seen.remove(s[left])
            left += 1
        seen.add(s[right])
        if right-left+1 > max_len:
            max_str = s[left:right+1]
        max_len = max(max_len, right-left+1)
        
    return max_str
    
```

```python
s = "abcdeacbed"
print_longest_string(s)
```

# Fixed size sliding window 
Given an integer array nums and an integer k, find the maximum sum of any subarray of size k.

nums = [2, 1, 5, 1, 3, 2]

k = 3

Subarrays of size 3:

[2, 1, 5] → 8

[1, 5, 1] → 7

[5, 1, 3] → 9

[1, 3, 2] → 6

```python
from collections import deque

def maximum_subarray(nums, k):
    sum_window = 0
    left = 0
    max_sum = float("-inf")
    max_subarray = []
    for right in range(len(nums)):
        sum_window += nums[right]
        if right-left+1 == k:
            if sum_window > max_sum:
                max_subarray = nums[left:right+1]
                max_sum = sum_window
            sum_window -= nums[left]
            left += 1
    return (max_subarray, max_sum)
```

```python
nums = [2,1,5,1,3,2]
maximum_subarray(nums, 3)
```

Assume n = length of array

- Time complexity: O(n) because each character enters and leaves the window at most once 
- Space complexity: O(k)
