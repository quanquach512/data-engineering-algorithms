# Valid parentheses 

"()[]{}" -> True

"(]" -> False

"([)]" -> False

"{[]}" -> True

```python
def is_valid(s):
    stack = []
    pairs = {")":"(","]":"[", "}":"{"}

    for c in s:
        if c in "{[(":
            stack.append(c)
        elif not stack or stack[-1] != pairs[c]:
            return False
        else:
            stack.pop()
    return len(stack) == 0
```

```python
print(is_valid("()[]{}"))

print(is_valid("(]"))
print(is_valid("([)]"))

print(is_valid("{[]}"))
```

- Time complexity: O(n)
- Space Complexity: O(n)
